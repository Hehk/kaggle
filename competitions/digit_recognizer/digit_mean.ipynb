```python
import torch
import numpy
```

```python
import pandas as pd
train_csv = pd.read_csv('./data/train.csv')
```

```python
train = torch.tensor(train_csv.values)
```

```python
train.shape
train[:28, :28].shape
```

```python
first_digit = train[0, 1:].reshape(28, 28)
first_digit.shape
```

```python
import matplotlib.pyplot as plt

plt.imshow(first_digit, cmap='gray')
```

```python
def show_digit(n):
    plt.imshow(train[n, 1:].reshape(28, 28), cmap='gray')
    plt.title(train[n, 0].item())
    plt.show()
```

```python
for n in range(10):
    show_digit(n)
```

```python
digits = train[:, 1:].reshape(42000, 28, 28)
avg_digits = torch.empty(10, 28, 28)
for i, x in enumerate(digits[:5]):
  label = train[i, 0]
  avg_digits[label.item()] += x
avg_digits
```
