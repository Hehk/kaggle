```python
import torch
class Dataset(torch.utils.data.Dataset):
  def __init__(self, df):
    self.date = df['date']
    self.store_nb = pd.get_dummies(df['store_nbr'])
    self.family = pd.get_dummies(df['family'])
    self.y = df['sales']
  
  def __len__(self):
    return len(self.date)
  
  def __getitem__(self, idx):
    # date = torch.tensor(self.date.iloc[idx])
    store_nb = torch.from_numpy(self.store_nb.iloc[idx].values).float()
    family = torch.from_numpy(self.family.iloc[idx].values).float()
    return [torch.cat((store_nb, family)), self.y.iloc[idx]]
```

```python
data = pd.read_csv('./data/train.csv')
dataset = Dataset(data)
dataset.__getitem__(5)[0]
```

```python
from torch import nn

# Starting with a model that ignores the dates
class Model(nn.Module):
  def __init__(self, n):
    super().__init__()
    self.layers = nn.Sequential(
      nn.Linear(n, n),
      nn.ReLU(),
      nn.Linear(n, n),
      nn.ReLU(),
      nn.Linear(n, 1),
      nn.ReLU()
    )
  
  def forward(self, x):
    return self.layers(x)
```

```python
dataset.__getitem__(0)[0]
```

```python
from torch.utils.data import DataLoader

train = DataLoader(dataset, batch_size=64, shuffle=True)

size = dataset.__getitem__(0)[0].shape[0]
model = Model(size)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_f = nn.MSELoss()

for epoch in range(10):
  running_loss = 0.0
  for i, data in enumerate(train, 0):
    inputs, labels = data
    
    optimizer.zero_grad()
    outputs = model(inputs.float())
    loss = loss_f(outputs, labels.float())
    loss.backward()
    optimizer.step()

    running_loss += loss.item()
    if i % 2000 == 1999:
      print(f'[{epoch + 1}, {i + 1:5d}] loss: {running_loss / 2000:.3f}')
      running_loss = 0.0

```
